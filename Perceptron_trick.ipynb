```python
class MyPerceptron:
  def __init__(self, learning_rate=0.01, epoch=100):
    self.intercept_ = 0
    self.coef_ = None
    self.lr = learning_rate
    self.epoch = epoch

  def step_function(self, z):
    return np.where(z >= 0, 1, 0)

  def fit(self, X_train, y_train):
    X_train = np.insert(X_train, 0, 1, axis=1)
    weights = np.ones(X_train.shape[1])

    for i in range(self.epoch):
        j = np.random.randint(0, X_train.shape[0])
        y_hat = self.step_function(np.dot(X_train[j], weights))
        weights = weights + self.lr * (y_train[j] - y_hat) * X_train[j]

    self.intercept_ = weights[0]
    self.coef_ = weights[1:]

    print(self.intercept_, self.coef_)

  def predict(self, X_test):
    linear_output = np.dot(X_test, self.coef_) + self.intercept_
    return self.step_function(linear_output)
```

```python
mp = MyPerceptron(learning_rate=0.1, epoch=1000000)
```

```python
mp.fit(X_train, y_train)
```

```python
mp.predict(X_test)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, mp.predict(X_test))
```
